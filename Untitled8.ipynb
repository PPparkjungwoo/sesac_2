```python
import networkx as nx
import matplotlib.pyplot as plt
import time
import random

# 도로 네트워크 그래프 생성
G = nx.Graph()

# 노드 추가 (교차로)
nodes = ['A', 'B', 'C', 'D', 'E']
G.add_nodes_from(nodes)

# 엣지 추가 (도로) 및 가중치 (거리) 설정
edges = [
    ('A', 'B', 5),
    ('A', 'C', 10),
    ('B', 'C', 3),
    ('B', 'D', 2),
    ('C', 'E', 4),
    ('D', 'E', 1)
]
G.add_weighted_edges_from(edges)

# 초기 침수된 도로 정보
flooded_roads = {
    ('A', 'C'): 0,  # 0cm 초기 상태
    ('C', 'E'): 0   # 0cm 초기 상태
}

# 특정 높이 이상의 빗물이 차올랐을 때 도로를 침수된 것으로 간주
threshold = 12  # cm

# 침수된 도로를 피하는 경로 계산 함수
def find_safe_path(graph, source, target, flooded_roads, threshold):
    G_safe = graph.copy()

    # 침수된 도로를 제거
    for road, water_level in flooded_roads.items():
        if water_level >= threshold:
            if G_safe.has_edge(*road):
                G_safe.remove_edge(*road)

    # 최단 경로 계산
    try:
        path = nx.shortest_path(G_safe, source=source, target=target, weight='weight')
        distance = nx.shortest_path_length(G_safe, source=source, target=target, weight='weight')
        return path, distance
    except nx.NetworkXNoPath:
        return None, float('inf')

# 시각화 함수
def visualize_graph(G, flooded_roads, threshold, safe_path=None):
    plt.figure(figsize=(10, 6))
    pos = nx.spring_layout(G)  # 노드 위치 설정
    nx.draw(G, pos, with_labels=True, node_size=700, node_color="lightblue", font_size=14, font_weight="bold")

    # 엣지 라벨 (거리) 추가
    edge_labels = {edge: f"{G.edges[edge]['weight']}m" for edge in G.edges}
    nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_color='green')

    # 침수된 도로 강조
    flooded_edges = [edge for edge, level in flooded_roads.items() if level >= threshold]
    nx.draw_networkx_edges(G, pos, edgelist=flooded_edges, edge_color='red', width=2)

    # 안전한 경로 강조
    if safe_path:
        safe_edges = list(zip(safe_path[:-1], safe_path[1:]))
        nx.draw_networkx_edges(G, pos, edgelist=safe_edges, edge_color='blue', width=2)

    plt.title("Navigation with Flooded Roads")
    plt.show()

# 출발지와 목적지 설정
source = 'A'
target = 'E'

# 드론으로부터 실시간 정보를 받아와서 업데이트하는 시뮬레이션
def drone_simulation(G, flooded_roads, threshold, source, target, iterations=10, delay=2):
    for _ in range(iterations):
        # 드론으로부터 실시간 침수 정보 업데이트 (랜덤값으로 시뮬레이션)
        for road in flooded_roads.keys():
            flooded_roads[road] = random.randint(0, 20)

        # 안전한 경로 찾기
        safe_path, safe_distance = find_safe_path(G, source, target, flooded_roads, threshold)

        # 그래프 시각화
        visualize_graph(G, flooded_roads, threshold, safe_path)

        # 경로 출력
        if safe_path:
            print(f"안전한 경로: {' -> '.join(safe_path)} (거리: {safe_distance})")
        else:
            print("안전한 경로를 찾을 수 없습니다.")

        # 지연 시간 대기 (실시간 업데이트 시뮬레이션)
        time.sleep(delay)

# 드론 시뮬레이션 실행
drone_simulation(G, flooded_roads, threshold, source, target)
```
